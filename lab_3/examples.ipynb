```python
import requests
from bs4 import BeautifulSoup
```

```python
# Задаем ссылку на статью, которую нужно распарсить
article_url = 'http://www.e1.ru/news/spool/news_id-450001-section_id-151.html'
r = requests.get(article_url)
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
# как искать в супе определенные теги или теги с заданными атрибутами
titleTag = soup.html.head.title
print(titleTag)
```

```python
# Выводим заголовок статьи
print(titleTag.string)
```

```python
# Выводим содержимое статьи
print(soup.get_text())
```

```python
# Содержимое мета полей
for meta in soup.find_all('meta'):
    print(meta.get('content'))
```

```python
# нахождение ссылок
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
# нахождение изображений
for img in soup.find_all('img'):
    print(img.get('src'))
```
